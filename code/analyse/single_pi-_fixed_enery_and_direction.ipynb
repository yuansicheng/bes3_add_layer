```python
import os, sys
import pandas as pd 
import numpy as np
from glob import glob
import matplotlib.pyplot as plt

this_path = os.path.abspath('.')
if this_path not in sys.path:
    sys.path.append(this_path)

from single_pp_fixed_enery_and_direction_func import root2Df
```

## 1. 读入数据、拟合

```python
def getSummary(root_path, kal=False):
    df_list = []
    index = []
    for d in os.listdir(root_path):
        index.append(float(d.split('_')[-1]))
        try:
            root_files = glob(os.path.join(root_path, d, '*.root'))       
            df_list.append(root2Df(root_files, kal=kal))
        except Exception as e:
            print(e)
            df_list.append(pd.Series([np.nan]*5))

    df = pd.concat(df_list, axis=1).T
    df.index = index
    
    df = df.sort_index()

    return df
    
```

### 1.1 液氢

```python
root_path = os.path.join(this_path, '../../data/gen/single_pi-/p_0.3_theta_0.1_phi_0.5/add_lh')

df_lh = getSummary(root_path, kal=False)
df_lh_kal = getSummary(root_path, kal=True)
```

### 1.2 液氘

```python
root_path = os.path.join(this_path, '../../data/gen/single_pi-/p_0.3_theta_0.1_phi_0.5/add_ld')

df_ld = getSummary(root_path, kal=False)
df_ld_kal = getSummary(root_path, kal=True)
```

## 2. 可视化

### 2.1 使用MDC track

```python
data = {
    'add_lh': df_lh, 
    'add_ld': df_ld,
}
for column in df_lh.columns:
    if column.endswith('_err'):
        continue
    for k,v in data.items():
        if column+'_err' in df_lh.columns:
            v[column].plot(title=column, grid=True, figsize=(4,2), style='.-', xlabel='thickness / mm', yerr=v[column+'_err'], label=k)
        else:
            v[column].plot(title=column, grid=True, figsize=(4,2), style='.-', xlabel='thickness / mm', label=k)
    plt.legend()
    plt.show()
```

### 2.2 使用kalman track

```python
data = {
    'add_lh': df_lh_kal, 
    'add_ld': df_ld_kal,
}
for column in df_lh.columns:
    if column.endswith('_err'):
        continue
    for k,v in data.items():
        if column+'_err' in df_lh.columns:
            v[column].plot(title=column, grid=True, figsize=(4,2), style='.-', xlabel='thickness / mm', yerr=v[column+'_err'], label=k)
        else:
            v[column].plot(title=column, grid=True, figsize=(4,2), style='.-', xlabel='thickness / mm', label=k)
    plt.legend()
    plt.show()
```
